# pyplot Tutorial

#### matplotlib.pyplot  
is a collection of command style that make matplotlib work like MATLAB.

Each pyplot function makes some changes to a figure: e.g., creates a figure, creates a plotting area in a figure, plot some lines in a plotting area, decorates the plot with labels etc

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,6,4])
plt.ylabel("Numbers")
plt.xlabel("Indeces")
plt.title("Myplot")
plt.show()
```

```python
plt.plot([1,4,9],[1,16,81])
plt.ylabel("Numbers")
plt.xlabel("Squares")
plt.title("Myplot")
plt.grid()  #grid on
plt.show()
```

```python
plt.plot([1,3,9],[1,16,81], "ro")    #r means red; o means dot
plt.grid()
plt.show()
```

```python
import numpy as np
t= np.arange(0,4,0.2)

plt.plot(t,t**2,    'b--' ,    label='^2')      #b--: blue ---
plt.plot(t,t**2.2,  'rs'  ,   label='^2.2')    #rs: red  square
plt.plot(t,t**2.5,  'g^'  ,   label='^2.5')   #g^: green triangle

plt.title("plot3")
plt.xlabel("t")
plt.ylabel("t^2, t^2.2,  t^2.5")


plt.legend()  #lagend can be under kind of symbolic scale
plt.grid()
plt.show()
```

# Controling Line Properties

###### use keyword args

```python
X=[1,2,3,4]
y=[1,4,9,16]   #Numpy array doesn't work with this function

plt.plot(X, y, linewidth=5.02)
plt.show()
```

Note: In  documentation for plot() functions there are too many partameters one can learn;  one of them are linewidth

```python
#setp: set-properties (properties ko set karna)


x1= [1,2,3,4]
y1= [1,4,9,16]
x2= [1,2,3,4]
y2= [1,3,1,5]

lines= plt.plot(x1,y1,x2,y2)
plt.grid()



#setp: set-properties (properties ko set karna) 

plt.setp(lines[0], color='r', linewidth=5.5)        #Python Style
plt.setp(lines[1],'color','g', 'linewidth',8.5)    #MATLAB Style
```

### Working with multiple figures and axes 
(Multiple plots)

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

t1 = np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0,5.0,0.02)

plt.figure(1)
plt.subplot(212)    #212 : 2 raw, 1 column, figure-2
plt.grid()
plt.plot(t1,f(t1), 'g^')     #g^: green trangle

plt.subplot(211)    #212 : 2 raw, 1 column, figure-1
plt.grid()
plt.plot(t2,np.cos(2*np.pi*t2), 'rs')   #rs: red square
plt.show()

#try same function second subplot by defining function
```

```python
#Exercise-1   playing with a numbers 

def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

def g(t):
    return np.cos(2*np.pi*t)

def h(t):
    return np.tan(2*np.pi*t)
def q(t):
    return np.exp(2*np.pi*t)


t1 = np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0,5.0,0.02)

plt.figure(1)
plt.subplot(212)    
plt.grid()
plt.plot(t1,f(t1), 'g^')    

plt.subplot(211)    
plt.grid()
plt.plot(t2,g(t2), 'rs')   
plt.show()

#*********Figure2***************#
plt.figure(2)
plt.subplot(221)    
plt.plot(t1,q(t1), 'rs')

plt.subplot(222)
plt.plot(t1,q(t1), 'gs')

plt.subplot(223)            #subplot(raw, column, position-3rd)
plt.plot(t1,h(t1), 'y^')

plt.subplot(224)
plt.plot(t2,f(t2), 'bs')
plt.grid()
plt.show()


```
