# Concatenate Data Frames 

```python
import pandas as pd     

#creating df using dictionary{} 
#Extra-notes:  dictonary is pythonspecfcterm but in Ds-algo it's called Hashmap
india_weather = pd.DataFrame({
    'city':['mumbai','delhi',' banglore'],
    'tempreature':[32,45,30],
    'humidity':[80,60,78]
})

india_weather
```

```python
us_weather = pd.DataFrame({
    'city':['Newyork','chicago','banglore'],
    'tempreature':[21,14,35],
    'humidity':[68,65,30]
})
us_weather
```

```python
#concate two dataframes
df = pd.concat([india_weather,us_weather])
#df = pd.concat([india_weather,us_weather], ignore_index=True)    #check this
df
```

```python
#Row-wise concatination
df = pd.concat([india_weather,us_weather],axis=1)
df
```
